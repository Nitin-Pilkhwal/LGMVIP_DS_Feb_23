## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Loading dataset

```python
%%capture --no-display
data = pd.read_csv("globalterrorismdb_0718dist.csv", encoding = 'ISO-8859-1')
```

## Feature Selection

```python
df = data[["iyear", "imonth", "iday", "country_txt", "provstate", "region_txt", 
           "longitude", "latitude", "attacktype1_txt", "target1", "nkill", "nwound", 
           "summary", "gname", "targtype1_txt", "weaptype1_txt", "motive"]]
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

## Handling missing data 

```python
df.duplicated().value_counts()
```

```python
%%capture --no-display
df.drop_duplicates(subset =None, keep='first', inplace = True)
df
```

```python
df.isnull().sum()
```

```python
%%capture --no-display
df['nkill'].fillna(0,inplace=True)
df['nwound'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

## Important Observation

```python
print("Country with most attack",df['country_txt'].value_counts().idxmax())
print("Year with most attack",df['iyear'].value_counts().idxmax())
print("Month with most attack",df['imonth'].value_counts().idxmax())
print("Day with most attack",df['iday'].value_counts().idxmax())
print("Region with most attack",df['region_txt'].value_counts().idxmax())
print("Most common type of weapon used in attack",df['weaptype1_txt'].value_counts().idxmax())
```

## Exploratory Data Analysis

```python
df.hist(figsize=(20,10))
```

```python
df.describe()
```

## Correlation Analysis

```python
%%capture --no-display
plt.figure(figsize=(5,5))
sns.heatmap(np.round(df.corr(),2),annot=True,cmap='BuPu')
```

```python
x_year = df['iyear'].unique()
y_year = df['iyear'].value_counts().sort_index()
plt.figure(figsize=(25,10))
sns.barplot(x=x_year,y=y_year,palette="rocket_r")
plt.xlabel("Attack Year")
plt.ylabel("Number of Attacks each year")
plt.title('Number of Attacks yearwise')
plt.show()
```

### Observation
Maximum number of attacks are in year 2014 and after that it gradually decreases yearwise.

```python
plt.subplots(figsize=(20,6))
sns.barplot(x=df['country_txt'].value_counts()[:30].index,y=df['country_txt'].value_counts()[:30].values,palette='Blues_r')
plt.title('Top 30 Countries that are Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

### Observation
<li>Country with most attacks is <b>Iraq</b> followed by Pakistan and Afganistan</li>
<li>Country with least number of attacks is <b>Nicaragua</b></li>

```python
plt.subplots(figsize=(30,10))
sns.barplot(x=df['region_txt'].value_counts().index,y=df['region_txt'].value_counts().values,palette='YlOrBr_r')
plt.title("Number of death region wise")
plt.xlabel('Region Name')
plt.ylabel('Count')
plt.show()
```

### Observation
<li>Most attacks are in <b>Middle East & North Africa</b> region.</li>
<li>Minimum attacks are in <b>Australasia & Oceansia</b> region.</li>

```python
plt.subplots(figsize=(20,15))
sns.countplot(y=df['targtype1_txt'],palette='viridis')
plt.xlabel('State Name')
plt.ylabel('Count')
plt.show()
```

### Observation
<li>Majority of the terrorists targets the Private Citizens and their properties and Military.</li>

```python
plt.subplots(figsize=(15,5))
sns.barplot(x=df['weaptype1_txt'].value_counts().index,y=df['weaptype1_txt'].value_counts().values,palette='Greens_r')
plt.title('Most Common Attack Type')
plt.xlabel('Attack Type')
plt.xticks(rotation = 90)
plt.ylabel('Count')
plt.show()
```

### Observation
<li>Most common attack type is <b>Explosives</b></li>

```python
plt.figure(figsize=(10,10))
plt.pie(df['gname'].value_counts()[:15].values,labels=df['gname'].value_counts()[:15].index,autopct='%1.2f%%')
plt.title('Top 10 Terrorist organization')
plt.legend(bbox_to_anchor=(1.0,1.0))
plt.show()
```

```python
fig=plt.figure(figsize=(12,3))
sns.barplot(x = df['gname'].value_counts()[1:11].values, y = df['gname'].value_counts()[1:11].index, palette='rocket')
plt.xlabel("No. of Attacks ")
plt.ylabel("Terrorist Organization")
plt.title("Most Active Terrorist Organization")
plt.show()
```

### Observation
<li>Taliban is the most active terrorist organisation followed by ISIL, Shining Path and so on.
</li>

```python
%%capture --no-display
Casualities = df.groupby(['country_txt'], dropna = False)['nkill','nwound'].sum()
Casualities['Casualities'] = Casualities.nkill + Casualities.nwound
Casualities_sorted = Casualities.sort_values('Casualities', ascending = False)
Casualities_sorted['country_txt'] = Casualities_sorted.index

fig=plt.figure(figsize=(20,3))
sns.barplot(x = Casualities_sorted['country_txt'][:20], y = Casualities_sorted['Casualities'][:20], palette='viridis')
plt.xlabel("Countries ")
plt.ylabel("Number of deaths")
plt.xticks(rotation = 90)
plt.title("Countries with maximum number of casualities")
plt.show()
```

### Observation
<li>Iraq has the highest number of casualities till date, followed by Afghanistan, Pakistan and so on. Thailand has comapratively lesser number of casualities.
</li>

```python
Attacks_2014 = df[df.iyear == 2014]
fig=plt.figure(figsize=(17,3))
sns.barplot(x = Attacks_2014['country_txt'].value_counts()[:20].index, y = Attacks_2014['country_txt'].value_counts()[:20].values, palette='viridis')
plt.xlabel("Countries ")
plt.ylabel("Number of attacks")
plt.xticks(rotation = 90)
plt.title("Most Attacked Countries in 2014")
plt.show()
```

### Observation
<li>
Iraq was most attacked country in the year 2014, followed by Pakistan.
</li>

```python
fig=plt.figure(figsize=(10,3))
sns.barplot(x = Attacks_2014['region_txt'].value_counts()[:20].index, y = Attacks_2014['region_txt'].value_counts()[:20].values, palette='magma')
plt.xlabel("Region")
plt.ylabel("Number of attacks")
plt.xticks(rotation = 90)
plt.title("Most Attacked Region in 2014")
plt.show()
```

### Observation
<li>
Middle East & North Africa was most attacked region, followed by South Asia.
</li>

```python
pd.crosstab(df.iyear,df.region_txt).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.xlabel('Years')
plt.show()
```

# Conclusion

<li>Heighest number of attacks took place in the year 2014 and then decreased gradually.
<li>Country that is most affected by these attacks are Iraq, Pakistan and Afghanistan.
<li>The most affected city was Baghdad.
<li>Middle East & North America is the most affected region.
<li>The most peaceful region is Australia & Oceania with minimum number of terrorist attacks.
<li>Majority of the terrorists targets the Private Citizens and their properties and followed by Military, Police, Government and so on.
<li>Taliban is the most active terrorist organisation after 2012.
